```python
import pandas as pd
df=pd.read_excel("British Airways Summer Schedule Dataset - Forage Data Science Task 1.xlsx")
df
```

```python
df["TIME_OF_DAY"].unique()
df["HAUL"].unique()
df["ARRIVAL_REGION"].unique()
```

```python
df["TOTAL_PASSENGERS"]= df["FIRST_CLASS_SEATS"] + df["BUSINESS_CLASS_SEATS"] + df["ECONOMY_SEATS"]
df["TIER1S"]=df["TIER1_ELIGIBLE_PAX"]
df["TIER2S"]=df["TIER2_ELIGIBLE_PAX"]
df["TIER3S"]=df["TIER3_ELIGIBLE_PAX"]
df["Total"]=df["TOTAL_PASSENGERS"]+df["TIER1_ELIGIBLE_PAX"]+df["TIER2_ELIGIBLE_PAX"]+df["TIER3_ELIGIBLE_PAX"]
grouped=df.groupby(["TIME_OF_DAY", "HAUL", "ARRIVAL_REGION"])["Total"].sum()
grouped
```

```python
df["Total"]=df["TOTAL_PASSENGERS"]+df["TIER1_ELIGIBLE_PAX"]+df["TIER3_ELIGIBLE_PAX"]
df["TIER1_ELIGIBLE_PCT"]=(df["TIER1_ELIGIBLE_PAX"]/df["Total"])*100
df["TIER2_ELIGIBLE_PCT"]=(df["TIER2_ELIGIBLE_PAX"]/df["Total"])*100
df["TIER3_ELIGIBLE_PCT"]=(df["TIER3_ELIGIBLE_PAX"]/df["Total"])*100



grouped=df.groupby(["TIME_OF_DAY", "HAUL", "ARRIVAL_REGION"])[["Total","TIER1_ELIGIBLE_PCT", "TIER2_ELIGIBLE_PCT", "TIER3_ELIGIBLE_PCT", "Total"]
].mean()
grouped=grouped.round(4)
grouped
grouped.to_excel("BA_Lounge_Lookup_Table.xlsx")
```
